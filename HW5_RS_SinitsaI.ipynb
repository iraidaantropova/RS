Методы коллаборативной фильтрации предполагают, что интересы пользователей представлены оценками, которые они дают объектам после просмотра, покупки и т.д. Основная идея данных методов заключается в сравнении между собой интересов различных пользователей или объектов на основе этих оценок. При этом никакой дополнительной информации о самих пользователях и объектах не используется.
Коллаборативная фильтрация – это метод рекомендации, при котором анализируется только реакция пользователей на объекты: оценки, которые выставляют пользователи объектам. Оценки могут быть как явными (пользователь явно указывает, на сколько «звездочек» он оценивает объект), так и неявными (например, количество просмотров одного ролика). Чем больше оценок собирается, тем точнее получатся рекомендации. Получается, что пользователи помогают друг другу в фильтрации объектов. Поэтому такой метод называется также совместной фильтрацией.

Крупные системы рекомендаций (СР), которые используют различные данные об объектах, пользователях, связях между ними, представляют собой сложную гибридную систему. Она является объединением различных подходов и алгоритмов: коллаборативной фильтрации, анализа содержимого, анализа социальных связей и связей между объектами, и т.д.

Приводится оценка следующих моделей на данных имитирующих warm-start и cold-start:

MF модель обычной матричной факторизации со сдвигами (bias) пользователей и предметов и с сигмоидной функцией.

LSI-LR модель, основанная на содержании-контенте. Извлекаются скрытые темы из матрицы item-фичей посредством латентной семантической индексации и представляются item'ы как линейные комбинации латентных тем. Затем создаютя отдельные модели логистической регрессии (LR) для каждого пользователя. В отличие от модели Light FM, которая использует колаборативные данные для создания своего латентного представления, LSI-LR основана исключительно на факторизации матрицы контента.

LSI-UP гибридная модель, представляющая пользовательские профили (UP) в виде линейных комбинаций контентных векторов item'ов, с последующим применением LSI к результирующей матрице для получения латентных представлений user'ов и item'ов.

LightFM (tags) модель LightFM только с фичами основанными на тегах.

LightFM (tags + ids) модель LightFM с фичами на тегах и индикаторных item фичах (ids)

LightFM (tags + about) модель Light FM с использованием как item, так и user фичей. Фичи пользователей можно получить только в датасете CrossValidated из раздела "About Me"
